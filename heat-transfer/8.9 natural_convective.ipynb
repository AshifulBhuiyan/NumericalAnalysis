# Natural Convective Heat Transfer Across a Rectangular Enclosure

## Examples:
- heat transfer between panes of glass in a double pane window
- heat transfer between collector plate and the glass cover in solar collector
- basic electronics

- AB is hot wall, kept at high temp $T_H$
- CD is cold wall, kept at low temp $T_C$
-fluid moves inside due to temperature gradient

To simplify the problem, we can make assumptions about temperature on vertical (end walls)
- BC and DA can be adiabatic (no heat transfer)
    - low thermal conductivity
- perfectly conducting (linear relation for $T_H$ to $T_C$)
    - made of material with high thermal conductivity

## Assumptions:
- flow in enclosure is steady and remains laminar
- flow is two-dimensional
- modelled by Boussinesq approximation:

$$
\frac{\partial u}{\partial x} + \frac{\partial v}{\partial y}= u_x +u_y= 0
$$
$$
u\cdot u_x +v\cdot u_y = \frac{1}{\rho}p_x + \nu(u_{xx}+u_{yy})+\beta g(T-T_1) \cos \phi
$$
$$
u\cdot u_x +v\cdot u_y = \frac{1}{\rho}p_y + \nu(u_{xx}+u_{yy})+\beta g(T-T_1) \sin \phi
$$
- these are Navier-Stokes in $x,y$ directions.
- LHS is the inertial (convective)acceleration
- $-\frac{-1}{\rho}p_x$ is the pressure gradient force (in $x$-dir)
- $\nu(u_{xx}+u_{yy})$ viscous diffusion (momentum spreading due to viscosity)
- $\beta g(T-T_1) \sin \phi$ is the buoyancy force component. Comes from Boussinesq approximation, where warmer fluid lighter and rises.
- $\phi$ is angle the enclosure makes with the vertical

$$
u T_x +v T_y = \left( \frac{k}{\rho c_p}\right) \left( T_{xx} +T_{yy}\right)
$$

- LHS is advection of heat (how fluid moves the heat around)
- RHS is thermal diffusion (heat conduction)
- $k$ = thermal conductivity
- $c_P=$ specific heat at constant pressure
- $\rho=$ density
- $\dfrac{k}{\rho c_P} = $ thermal diffusivity $\alpha$
