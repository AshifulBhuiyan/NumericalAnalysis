```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
#Grid:
def grid(a,b,c,d,n,m):
    h = (b - a) / n
    k = (d - c) / m
    x=np.linspace(a,b,n+1)
    y=np.linspace(c,d,m+1)
    
    return x, y, h, k

#Boundary
def boundary(x,y, gxc, gxd, gya, gyb, n,m):
    w=np.zeros((n+1,m+1))
    for i in range(n+1):
        w[i,0] = gxc(x[i]) #left boundary
        w[i,m] = gxd(x[i]) #right boundary
    for j in range(m+1):
        w[0,j] = gya(y[j]) #lower boundary
        w[n,j] = gyb(y[j]) #upper boundary
    return w
```

```python
#following textbook

def poissonDiff(x,y,h,k,n,m, f, TOL, N):
    lam = (h**2)/k**2
    mu = 2*(1+lam)
    l = 1
    w = np.zeros((n+1, m+1))
    w[:,0], w[0,:], w[:,m], w[n,:] = boundary_conditions(x, y)
    #w= np.transpose(w)
    print(np.array_str(w, precision=2, suppress_small=True))
   
    while l <= N: #Gauss-Seidel Iterations
        z = (-h**2 * f(x[0],y[m-1]) + w[0,m-1] + lam * w[0,m] + lam* w[0,m-2] + w[1,m-1])/mu
        norm = np.abs(z-w[0,m-1])
        w[0,m-1] = z
        #print(np.array_str(w, precision=2, suppress_small=True))
        for i in range(1,n-2):
            z = (-h**2 * f(x[i],y[m-1]) + lam*w[i,m] + w[i-1,m-1] + w[i+1,m-1] + lam * w[i,m-2])/mu
            if np.abs(w[i,m-1] - z) > norm:
                norm = np.abs(w[i,m-1] - z)
           #     print(f"step 1 + {norm}")
            w[i,m-1] = z
        z = (-h**2 * f(x[i-1],y[m-1]) + w[n,m-1]+ lam*w[n-1,m] + w[n-2,m-1] + lam * w[n-1,m-2])/mu #step 9
        if np.abs(w[n-1,m-1] - z) > norm:
            norm = np.abs(w[n-1,m-1] - z)
        #    print(f"step 1.5 + {norm}")
        w[n-1,m-1] = z
        for j in range(m-2,0,-1): #step 10
            z = (-h**2*f(x[0], y[j]) + w[0,j] + lam*(w[0,j+1] + w[0,j-1]) + w[1,j] )/mu
            if np.abs(w[0,j] - z) > norm:
                norm = np.abs(w[0,j] - z)
            w[0,j] = z
      #      print(np.array_str(w, precision=2, suppress_small=True))
      #      print(f"step 2 + {norm}")
            for i in range(1,n-1): #step 12
                z = (-h**2*f(x[i], y[j]) + w[i-1,j] + lam*w[i,j+1]  + w[i+1,j] + lam*w[i,j-1] )/mu
                if np.abs(w[i,j] - z) > norm:
                    norm = np.abs(w[i,j] - z)
                    #print(f"step 3 + {norm}")
                w[i,j] = z
         #       print(np.array_str(w, precision=2, suppress_small=True))
            z = (-h**2 * f(x[n-1], y[j]) + w[n,j] + w[n-2,j] + lam*w[n-1,j+1]+lam*w[n-1,j-1])/mu #step 13
            if np.abs(w[n-1,j] - z) > norm:
                norm = np.abs(w[n-1,j]-z)
            w[n-1,j] = z
       #     print(np.array_str(w, precision=2, suppress_small=True))
       #     print(f"step 4 + {norm}")
        z = (-h**2*f(x[0],y[0])+w[0,0] + lam*w[0,0]+lam*w[0,1]+w[1,0])/mu #step 14
        if np.abs(w[1,1] - z) > norm:
            norm = np.abs(w[1,1]-z)
        #    print(f"step 5 + {norm}")
        w[1,1] = z
        #print(np.array_str(w, precision=2, suppress_small=True))
        for i in range(2,n-1): #step 15
            z = (-h**2 * f(x[i],y[0]) + lam* w[i,0]+w[i-1,0]+ lam*w[i,1]+w[i+1,0])/mu
            if np.abs(w[i,1]-z) > norm:
                norm = np.abs(w[i, 1 ]- z)
            w[i,1] = z
         #   print(f"step 6 + {norm}")
        z = (-h**2 * f(x[n-1], y[0])+w[n,0]+lam*w[n-1,0]+w[n-2,0] + lam * w[n-1,1])/mu #step 16
        if np.abs(w[n-1,0]-z) > norm:
            norm = np.abs(w[n-1,0] - z)
        #    print(f"step 7 + {norm}")
        w[n-1,1] = z
        if norm <= TOL:
            print("norm under TOL")
            print(np.array_str(w, precision=2, suppress_small=True))
            return w
        l = l+1
    print("max iteration exceeded")
    print(np.array_str(w, precision=2, suppress_small=True))
    return w
    
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Grid definition
def create_grid(a, b, c, d, n, m):
    h = (b - a) / n
    k = (d - c) / m
    x = np.linspace(a, b, n + 1)
    y = np.linspace(c, d, m + 1)
    return x, y, h, k

# Boundary conditions
def apply_boundary_conditions(w, x, y, leftB, rightB, lowerB, upperB):
    n, m = w.shape[0] - 1, w.shape[1] - 1
    for i in range(n + 1):
        w[i, 0] = leftB(y[i])  # Left boundary
        w[i, m] = rightB(y[i])  # Right boundary
    for j in range(m + 1):
        w[0, j] = lowerB(x[j])  # Lower boundary
        w[n, j] = upperB(x[j])  # Upper boundary
    return w

# Numerical solution for Poisson equation using Gauss-Seidel
def poisson_solver(x, y, h, k, f, leftB, rightB, lowerB, upperB, TOL, max_iter):
    n, m = len(x) - 1, len(y) - 1
    w = np.zeros((n + 1, m + 1))
    w = apply_boundary_conditions(w, x, y, leftB, rightB, lowerB, upperB)

    lam = h**2 / k**2
    mu = 2 * (1 + lam)
    
    for iteration in range(max_iter):
        norm = 0

        # Update interior points
        for i in range(1, n):
            for j in range(1, m):
                z = (-h**2 * f(x[i], y[j]) + w[i - 1, j] + w[i + 1, j] + lam * (w[i, j - 1] + w[i, j + 1])) / mu
                norm = max(norm, abs(w[i, j] - z))
                w[i, j] = z

        # Check convergence
        if norm < TOL:
            print(f"Converged after {iteration + 1} iterations.")
            return w

    print("Max iterations exceeded without convergence.")
    return w

# Example usage
def example():
    # Problem definition
    a, b, c, d = 0, 1, 0, 1  # Domain
    n, m = 10, 10  # Number of intervals
    TOL = 1e-6
    max_iter = 1000

    # Source function
    def f(x, y):
        return -2 * (x**2 + y**2)

    # Boundary conditions
    def leftB(y):
        return 0

    def rightB(y):
        return 0

    def lowerB(x):
        return x**2

    def upperB(x):
        return x**2 + 1

    # Create grid
    x, y, h, k = create_grid(a, b, c, d, n, m)

    # Solve
    w = poisson_solver(x, y, h, k, f, leftB, rightB, lowerB, upperB, TOL, max_iter)

    # Plot result
    X, Y = np.meshgrid(x, y)
    plt.figure(figsize=(8, 6))
    cp = plt.contourf(X, Y, w.T, 20, cmap="viridis")
    plt.colorbar(cp)
    plt.title("Numerical Solution of Poisson Equation")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

# Run example
if __name__ == "__main__":
    example()
```
